```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Subject eating
s1 = pd.read_csv('eating_subj1record-[2019.02.25-11.04.56].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s1.fillna(value=0, inplace=True)
s2 = pd.read_csv('eating_subj2record-[2019.02.25-11.24.56].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s2.fillna(value=0, inplace=True)
s3 = pd.read_csv('eating_subj3record-[2019.02.25-12.07.01].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s3.fillna(value=0, inplace=True)
s4 = pd.read_csv('eating_subj4record-[2019.02.25-12.24.38].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s4.fillna(value=0, inplace=True)
s5 = pd.read_csv('eating_subj5record-[2019.02.25-12.37.01].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s5.fillna(value=0, inplace=True)
s6 = pd.read_csv('eating_subj6record-[2019.02.26-17.48.02].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s6.fillna(value=0, inplace=True)
s7 = pd.read_csv('eating_subj7record-[2019.02.26-17.55.20].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s7.fillna(value=0, inplace=True)
s8 = pd.read_csv('eating_subj8record-[2019.02.26-18.33.44].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
s8.fillna(value=0, inplace=True)
```

```python
#Waldo baseline attention
w1 = pd.read_csv('waldo_subj1record-[2019.02.25-11.04.56].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w1.fillna(value=0, inplace=True)
w2 = pd.read_csv('waldo_subj2record-[2019.02.25-11.27.22].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w2.fillna(value=0, inplace=True)
w3 = pd.read_csv('waldo_subj3record-[2019.02.25-12.10.57].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w3.fillna(value=0, inplace=True)
w4 = pd.read_csv('waldo_subj4record-[2019.02.25-12.28.48].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w4.fillna(value=0, inplace=True)
w5 = pd.read_csv('waldo_subj5record-[2019.02.25-12.55.09].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w5.fillna(value=0, inplace=True)
w6 = pd.read_csv('waldo_subj6record-[2019.02.26-17.52.49].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w6.fillna(value=0, inplace=True)
w7 = pd.read_csv('waldo_subj7record-[2019.02.26-17.59.43].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w7.fillna(value=0, inplace=True)
w8 = pd.read_csv('waldo_subj8record-[2019.02.26-18.39.03].csv', header=0, usecols=['Time:512Hz', 'Epoch', 'Electrode', 'Event Id'])
w8.fillna(value=0, inplace=True)
```

```python
h = 512*15 #512Hz times n seconds to be dropped
cut_s1 = s1[h:] #new DataFrame omits the first 15 seconds of data
cut_s2 = s2[h:] 
cut_s3 = s3[h:]
cut_s4 = s4[h:] 
cut_s5 = s5[h:]
cut_s6 = s6[h:]
cut_s7 = s7[h:] 
cut_s8 = s8[h:]
```

### Subject 2 - s2

```python
Xcs2 = cut_s2['Time:512Hz'] #cutoff data X (time)
Ycs2 = cut_s2['Electrode'] #cutoff data Y (amplitude)
Zcs2 = cut_s2['Event Id'] #cutoff data Z (event amplitude)

Xcw2 = cut_w2['Time:512Hz'] #cutoff data X Waldo
Ycw2 = cut_w2['Electrode'] #cutoff data Y Waldo
```

```python
#plots subject data and waldo data
fig, (ax1) = plt.subplots(1, figsize=(35,20), sharex=True)
ax1.set_title('Subject 2')
ax1.plot(Xcs2, Ycs2)
ax1.plot(Xcw2, Ycw2, color='r')
ax1.set_ylabel('Amplitude')
ax1.set_ylim(-2000, 2000)
plt.rcParams.update({'font.size': 30})
plt.show()
```

```python
#first 5 second interval 15sec to 20sec
a = 15
b = 20
x = Xcs2.loc[512*a:512*b]
y = Ycs2.loc[512*a:512*b]
xw = Xcw2.loc[512*a:512*b]
yw = Ycw2.loc[512*a:512*b]

fig, (ax1) = plt.subplots(1, figsize=(35,20), sharex=True)
ax1.plot(x, y)
ax1.plot(xw, yw, color='r')
ax1.set_ylabel('Amplitude')
fig.subplots_adjust(hspace=0)
ax1.set_title('15-20 seconds')
ax1.set_ylim(-2000, 2000)
plt.rcParams.update({'font.size': 30})
plt.show()
#could be a bite event in between 17-18 seconds
#could be a bite event in between 19.5-20 seconds
```

```python
#2 second interval 17sec to 19sec
a = 17
b = 19
x = Xcs2.loc[512*a:512*b]
y = Ycs2.loc[512*a:512*b]
xw = Xcw2.loc[512*a:512*b]
yw = Ycw2.loc[512*a:512*b]

fig, (ax1) = plt.subplots(1, figsize=(35,20), sharex=True)
ax1.plot(x, y)
ax1.plot(xw, yw, color='r')
ax1.set_ylabel('Amplitude')
fig.subplots_adjust(hspace=0)
ax1.set_title('17-19 seconds')
ax1.set_ylim(-2000, 2000)
plt.rcParams.update({'font.size': 30})
plt.show()
```
