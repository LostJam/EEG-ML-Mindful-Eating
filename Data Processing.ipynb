```python
h = 512*15 #512Hz times n seconds to be dropped
cut_s1 = s1[h:] #new DataFrame omits the first 15 seconds of data
cut_s2 = s2[h:] 
cut_s3 = s3[h:]
cut_s4 = s4[h:] 
cut_s5 = s5[h:]
cut_s6 = s6[h:]
cut_s7 = s7[h:] 
cut_s8 = s8[h:]
```

### Subject 2 - s2

```python
Xs2 = s2['Time:512Hz'] #original data X (time)
Ys2 = s2['Electrode'] #original data Y (amplitude)

Xcs2 = cut_s2['Time:512Hz'] #cutoff data X (time)
Ycs2 = cut_s2['Electrode'] #cutoff data Y (amplitude)
Zcs2 = cut_s2['Event Id'] #cutoff data Z (event amplitude)
```

```python
#plots 2 graphs: 1.) original data 2.) data after 15 second cutoff
fig, (ax1, ax2) = plt.subplots(2, figsize=(35,40), sharex=True)
ax1.set_title('Subject 2')
ax1.plot(Xcs2, Ycs2)
ax2.set_xlabel('Time (shared x-axis)')
ax1.set_ylabel('Amplitude')
ax2.set_ylabel('Amplitude')
ax2.plot(Xcs2, Zcs2)
fig.subplots_adjust(hspace=0)
plt.rcParams.update({'font.size': 30})
plt.show()
```

```python
#first 10 second interval 15sec to 25sec
a = 25
b = 35
x = Xcs2.loc[512*a:512*b]
y = Ycs2.loc[512*a:512*b]
z = Zcs2.loc[512*a:512*b]

fig, (ax1, ax2) = plt.subplots(2, figsize=(35,20), sharex=True)
ax1.plot(x, y)
ax2.plot(x, z)
ax1.set_ylabel('Amplitude')
fig.subplots_adjust(hspace=0)
ax1.set_title('15-25 seconds')
ax1.set_ylim(-2000, 2000)
ax2.set_ylim(-2000, 2000)
ax2.set_ylabel('Amplitude')
fig.subplots_adjust(hspace=0)
plt.rcParams.update({'font.size': 30})
plt.show()
#  10 second interval gets really cramped to look at
```

```python
#first 5 second interval 15sec to 20sec
a = 15
b = 20
x = Xcs2.loc[512*a:512*b]
y = Ycs2.loc[512*a:512*b]

fig, (ax1) = plt.subplots(1, figsize=(35,20), sharex=True)
ax1.plot(x, y)
ax1.set_ylabel('Amplitude')
fig.subplots_adjust(hspace=0)
ax1.set_title('15-20 seconds')
ax1.set_ylim(-2000, 2000)
plt.rcParams.update({'font.size': 30})
plt.show()
#could be a bite event in between 17-18 seconds
#could be a bite event in between 19.5-20 seconds
```
